## Limpieza de variables

#### Año de ocurrencia

Notamos que la variable año de registro (*anio_regis*) tiene abreviados a 2 cifras los años 1990 a 1997.

Para la variable año de ocurrencia, vamos a considerar del mismo año las variables 90 y 1990, 91 y 1991 y así sucesivamente hasta el año 97 y 1997. Los demás valores no se considerarán.

```python
by_anio_regis = homicidios_vars.anio_regis.value_counts().sort_index()
#print(by_anio_regis)
by_anio_ocurr = homicidios_vars.anio_ocur.value_counts().sort_index()
#print(by_anio_ocurr)

homicidios_vars.loc[homicidios_vars.anio_ocur == 90, 'anio_ocur'] = 1990
homicidios_vars.loc[homicidios_vars.anio_ocur == 91, 'anio_ocur'] = 1991
homicidios_vars.loc[homicidios_vars.anio_ocur == 92, 'anio_ocur'] = 1992
homicidios_vars.loc[homicidios_vars.anio_ocur == 93, 'anio_ocur'] = 1993
homicidios_vars.loc[homicidios_vars.anio_ocur == 94, 'anio_ocur'] = 1994
homicidios_vars.loc[homicidios_vars.anio_ocur == 95, 'anio_ocur'] = 1995
homicidios_vars.loc[homicidios_vars.anio_ocur == 96, 'anio_ocur'] = 1996
homicidios_vars.loc[homicidios_vars.anio_ocur == 97, 'anio_ocur'] = 1997
#homicidios_vars.anio_ocur.loc[(homicidios_vars['anio_ocur'] == 90)] = 1990
#homicidios_vars['anio_ocur'] = np.where(homicidios_vars['anio_ocur'] == 90, 1990, homicidios_vars['anio_ocur'])

by_anio_ocurr = homicidios_vars.anio_ocur.value_counts().sort_index()
#print(by_anio_ocurr[105:135,])
```

#### Entidades de ocurrencia

Se muestra el porcentaje de valores faltantes para la variable entidad de ocurrencia por año de ocurrencia

```python
edo_ocur_missing = homicidios_vars.loc[homicidios_vars.ent_ocurr > 32].groupby(['anio_ocur']).size()
edo_ocur_missing = pd.DataFrame(edo_ocur_missing)
edo_ocur_missing.columns = ['#_missings']
edo_ocur_missing['%_missings'] = edo_ocur_missing['#_missings']/(by_anio_ocurr)
edo_ocur_missing.loc[1990:2018]
```

#### Localidad de ocurrencia

Se muestra el porcentaje de valores faltantes para la variable localidad de ocurrencia por año de ocurrencia.

```python
loc_ocur_missing = homicidios_vars.loc[(homicidios_vars.loc_ocurr.isna()) | (homicidios_vars.loc_ocurr == 99)].groupby(['anio_ocur']).size()
loc_ocur_missing = pd.DataFrame(loc_ocur_missing)
loc_ocur_missing.columns = ['#_missings']
loc_ocur_missing['%_missings'] = loc_ocur_missing['#_missings']/(by_anio_ocurr)
loc_ocur_missing.loc[1990:2018]
```
